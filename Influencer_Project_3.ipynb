```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Influencer.xlsx')
df
```

```python
def change(df,list1):
    for i in list1:
      df[i]=df[i].replace({"b":"1e9","k":"*1e3", "m":"*1e6"}, regex=True).map(pd.eval)
    return df
```

```python
change(df,['Followers','Avg. Likes','Posts','New Post Avg. Likes','Total Likes'])
```
